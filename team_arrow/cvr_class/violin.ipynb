```python
import pandas as pd 
import seaborn as sns
```

```python
df = pd.DataFrame(columns= ['candidates', 'choices', 'gamma', 'partisan'])
print(df)
```

```python
elections = pd.read_csv("election_table_2.csv")
d = {}
for i, row in elections.iterrows():
    partisan = None
    primary = False
    if row["partisan"] == '0':
        partisan = False
    elif row["partisan"] == '1':
        partisan = True 
    else:
        primary = True 
    
    if primary is False and row["candidates"] < 20:
        gamma = row["gamma mds"]
        choices = row["choices"]
        candidates = row['candidates']
        df.loc[len(df)] = [candidates, choices, gamma, partisan]



```

```python
print(df)
```

```python
sns.violinplot(data=df, x='choices', y='gamma', hue='partisan', split=True)
```

```python
sns.boxplot(data=df, x='choices', y='gamma', hue='partisan')
```

```python
sns.boxenplot(data=df, x='candidates', y='gamma', hue='partisan')
```
